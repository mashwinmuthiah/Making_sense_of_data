```python
import pandas as pd
from bs4 import BeautifulSoup
import requests 
from urllib.request import urlopen
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)"
html = urlopen(url).read()
```

```python
soup = BeautifulSoup(html ,"html.parser")
```

```python
soup.title
```

```python
table = soup.find_all('table')
#table[1]
table1 = {'Rank':[],'Country or Area':[],'UN continental region':[],'UN statistical region':[],'Population (1 July 2016)':[],'Population (1 July 2017)':[],'Change':[]}
```

```python
rows = table[1].find_all("tr")[2:]
#rows[0]
```

```python
for r in rows:
    td = r.find_all('td')
    table1['Rank'].append(td[0].string)
    table1['Country or Area'].append(td[1].string)
    table1['UN continental region'].append(td[2].string)
    table1['UN statistical region'].append(td[3].string)
    table1['Population (1 July 2016)'].append(td[4].string)
    table1['Population (1 July 2017)'].append(td[5].string)
    table1['Change'].append(td[6].string)
    
```

```python
df = pd.DataFrame(table1)
df
```

```python
df.to_csv("Word_population.csv",index=False)
```
